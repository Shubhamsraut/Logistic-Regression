```python
# import libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
# Load the data using seperator to convert text into different columns
df = pd.read_csv("bank-full.csv",sep = ';')
df
```

```python
# To convert object data type into numeric
# Custom Binary Encoding --> with o/p variables
df1['default'] = np.where(df1['default'].str.contains("yes"),1,0)
df1['housing'] = np.where(df1['housing'].str.contains("yes"),1,0)
df1['loan'] = np.where(df1['loan'].str.contains("yes"),1,0)
df1['y'] = np.where(df1['y'].str.contains("yes"),1,0)
```

```python
# month --> not a binary o/p variables -->brings many columns so --> 
order={'month':{'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}}
```

```python
df1 = df1.replace(order)
df1
```

```python
df1.info()
```

 ### 2.Model Building 

```python
# Dividing the data into input and output variables
x = pd.concat ([df1.iloc[:,0:11],df1.iloc[:,12:]],axis = 1) # concat --> add all needed columns and consider them in x
y = df1.iloc[:,11] # 11th column
```

```python
x
```

```python
y
```

##### classifier --> used to predict result into binary --> yes /no --> 0/1

```python
# Building a Logistic regression model
classifier = LogisticRegression() #  to predict the class (or category) of individuals based on one or multiple predictor variables (x)
classifier.fit(x,y)
```

### 3. Model Predictions

```python
# Prediction for all x data sets
y_pred = classifier.predict_proba(x)
y_pred
```

```python
y_pred = classifier.predict(x)
y_pred
```

### 5.ROC curve and AUC value

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt
```

```python
fpr, tpr, thresholds = roc_curve(y, classifier.predict_proba (x)[:,1])
auc = roc_auc_score(y, y_pred)

plt.plot(fpr, tpr, color='blue', label='logit model ( area  = %0.2f)'%auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()
```
